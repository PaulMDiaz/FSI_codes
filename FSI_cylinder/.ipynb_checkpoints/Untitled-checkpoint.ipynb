```python
from dolfin import *
import numpy as np
import pylab as plt
from mshr import *
from IPython.display import HTML

# Geometric parameters
L = 2.5 	#Channel length
H = 0.41	#Channel height
l = 0.35 	#Bar length
h = 0.02	#Bar height
N = 64
xw = h      #with about bar for refinement. 
x_bar = 0.6 - l

channel = Rectangle(Point(0, 0), Point(L, H))
cylinder = Circle(Point(0.2, 0.2), 0.05,50)

# this definition leaves bar tangent to circle with small gap
bar = Rectangle(Point(x_bar,0.19), Point(0.6, 0.19+h))

# Ensures overlap between bar and circle.
bar_2 = Rectangle(Point(x_bar-0.01,0.19), Point(0.6, 0.19+h))

# fluid domain
f_domain = channel - cylinder - bar_2

# structure domain
s_domain = bar

# total domain
domain = channel -cylinder -bar_2 +bar
# try self.domain = self.f_domain +self.s_domain

# Set structure as subdomain, first type for global mesh
domain.set_subdomain(1,s_domain)

mesh = generate_mesh(domain, N)

#cell_markers = CellFunction("bool", mesh, False)
#structure = AutoSubDomain(lambda x: x[0] >= x_bar and x[0] <= 0.6 and x[1] >= 0.19 and x[1] <= 0.19+h)
#structure.mark(cell_markers,True)
#mesh = refine(mesh, cell_markers)

# reset structure domain
domain.set_subdomain(1,s_domain)

mesh_f = SubMesh(mesh, 0)
mesh_s = SubMesh(mesh, 1)


cell_markers = CellFunction("bool", mesh_s, False)
structure = AutoSubDomain(lambda x: x[0] >= x_bar and x[0] <= 0.6 and x[1] >= 0.19 and x[1] <= 0.19+h)		#fsi = CompiledSubDomain('on_boundary && x[0] > x_bar + DOLFIN_EPS && x[0]< L - DOLFIN_EPS && x[1]>DOLFIN_EPS && x[1] < H-DOLFIN_EPS', L = self.L, H = self.H, x_bar = self.x_bar)
structure.mark(cell_markers,True)
mesh_s = refine(mesh_s, cell_markers)

cell_markers = CellFunction("bool", mesh, False)
struct_rect = AutoSubDomain(lambda x: x[0] >= x_bar -xw and x[0] <= 0.6 + xw and x[1] >= 0.19 -xw and x[1] <= 0.19+h+xw)
struct_rect.mark(cell_markers,True)
mesh = refine(mesh, cell_markers)

cell_markers = CellFunction("bool", mesh_f, False)
struct_rect = AutoSubDomain(lambda x: x[0] >= x_bar -xw and x[0] <= 0.6 + xw and x[1] >= 0.19 -xw and x[1] <= 0.19+h+xw)
structure_rect.mark(cell_markers,True)
mesh_f = refine(mesh_f, cell_markers)

V_space = VectorFunctionSpace(mesh_f, 'CG', 2)


HTML(X3DOM().html(mesh_f))
```
